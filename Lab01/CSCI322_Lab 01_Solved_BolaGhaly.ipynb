# Lab tasks

##### (1) fbs: The person's fasting blood sugar "fbs" (> 120 mg/dl, 1 = true; 0 = false)

* create new column 'bloodSugar' with values 'Yes' if fbs=1 or 'No' if fbs=0

```python
## your code here
df['bloodSugar'] = df['fbs'].apply(lambda x: 'Yes' if x == 1 else 'No')

df[['bloodSugar','fbs']]
```

##### (2) cp: The chest pain experienced (Value 1: typical angina, Value 2: atypical angina, Value 3: non-anginal pain, Value 4: asymptomatic)

* convert chest pain column values to their corresponding meanings

```python
df['cp'].unique()
```

```python
## your code here
meanings = {0: 'typical angina', 1: 'atypical angina', 2: 'non-anginal pain', 3: 'asymptomatic'}
df['cp'] = df['cp'].map(meanings)
df['cp']

```

##### (3) crosstab: attack and bloodSugar cloumns VS. gender and chestPain cloumns


```python
## your code here
table2 = pd.crosstab( index = [df['heart_attack'],df['bloodSugar']], columns = [df['sex'],df['cp']] , margins = True)
table2
```
