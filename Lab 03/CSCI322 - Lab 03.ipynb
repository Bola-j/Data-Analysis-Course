# data analysis steps
### We organized the data analysis process into five steps:
* <b>step 1: Ask questions</b>

Either you're given data and ask questions based on it, or you ask questions first and gather data based on that later.

* <b>Step 2: Wrangle data</b>

You get the data you need in a form you can work with in three steps: 
* 1. gather
* 2. assess
* 3. clean

You gather the data you need to answer your questions, assess your data to identify any problems in your data’s quality or structure, and clean your data by modifying, replacing, or removing data to ensure that your dataset is of the highest quality and as well-structured as possible.

* <b>Step 3: Perform EDA (Exploratory Data Analysis)</b>

You explore and then augment your data to maximize the potential of your analyses, visualizations, and models. Exploring involves finding patterns in your data, visualizing relationships in your data, and building intuition about what you’re working with. After exploring, you can do things like remove outliers and create better features from your data, also known as feature engineering.

* <b>Step 4: Draw conclusions (or even make predictions)</b>

This step is typically approached with machine learning or inferential statistics.

* <b>Step 5: Communicate your results</b>

You often need to justify and convey meaning in the insights you’ve found. Or, if your end goal is to build a system, you usually need to share what you’ve built, explain how you reached design decisions, and report how well it performs. There are many ways to communicate your results: reports, slide decks, blog posts, emails, presentations, or even conversations. Data visualization will always be very valuable.

# Data cleaning
Let's first practice handling missing values and duplicate data

```python
df = pd.read_excel('cancer_data.xlsx', sheet_name='cancer_data_means')
df.head()
```

```python
# Check for missing or null values
df.info()
```

```python
# Check for missing or null values using df.isna()
## your code here
df.isna().sum()
```

# Filling the missing values
[Handling Missing Values in Data Analysis](https://www.analyticsvidhya.com/blog/2021/10/handling-missing-value/)

```python
# use means to fill in missing values
## your code here


df['texture_mean'] = df['texture_mean'].fillna(df['texture_mean'].mean())
df['smoothness_mean'] = df['smoothness_mean'].fillna(df['smoothness_mean'].mean())
df['symmetry_mean'] = df['symmetry_mean'].fillna(df['symmetry_mean'].mean())
```

```python
# confirm your correction with info()
## your code here

df.isna().sum()
```

# Extra
* Fill missing values using SimpleImputer

```python
from sklearn.impute import SimpleImputer
df2 = pd.read_excel('cancer_data.xlsx', sheet_name='cancer_data_means')
print(df2.isna().sum())
print('DataFrame  after filling null values:')
imputer = SimpleImputer(strategy='median')
df2.loc[:, df.select_dtypes(include=np.number).columns] = imputer.fit_transform(df2.select_dtypes(include=np.number))
print(df2.isna().sum())
```

```python
# check for duplicates in the data
## your code here

df.duplicated().sum()
```

```python
# drop duplicates
## your code here

df = df.drop_duplicates()
```

```python
# confirm correction by rechecking for duplicates in the data
## your code here

df.duplicated().sum()
```

### Renaming Columns

```python
# remove "_mean" from column names
new_labels = []
for col in df.columns:
    if '_mean' in col:
        new_labels.append(col[:-5])  # exclude last 6 characters
    else:
        new_labels.append(col)

# new labels for our columns
new_labels
```

```python
# assign new labels to columns in dataframe
## your code here
df.columns = new_labels


# display first few rows of dataframe to confirm changes
df.head()
```

```python
# save this dataset for later
df.to_excel('cancer_data_cleaned.xlsx', index=False)
```
